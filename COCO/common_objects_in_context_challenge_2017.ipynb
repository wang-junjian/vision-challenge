# Common Objects in Context Challenge 2017

## 目录结构

* COCO2017
    * annotations_trainval2017
        * annotations `captions: 看图说话; instances: 目标实例; person_keypoints: 目标上的关键点;`
            - captions_train2017.json [格式](#annotations/annotations/captions_x.json)
            - captions_val2017.json
            - instances_train2017.json [格式](#annotations/annotations/instances_x.json)
            - instances_val2017.json
            - person_keypoints_train2017.json [格式](#annotations/annotations/person_keypoints_x.json)
            - person_keypoints_val2017.json
    * image_info_test2017
        * annotations
            - image_info_test2017.json [格式](#image_info/annotations/image_info_x.json)
            - image_info_test-dev2017.json
    * stuff_annotations_trainval2017
        * annotations
            - stuff_train2017.json
            - stuff_train2017_pixelmaps.zip
            - stuff_val2017.json
            - stuff_val2017_pixelmaps.zip
    * train2017 `训练图片集`
        - 000000000009.jpg
        - 000000000025.jpg
    * val2017 `验证图片集`
        - 000000000139.jpg
        - 000000000285.jpg
